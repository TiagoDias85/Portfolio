```python
import pandas as pd
```

```python
dados_csv = pd.read_csv("Registro_Vendas.csv")

display(dados_csv)
```

```python
total_venda_produto_A = dados_csv[dados_csv['Produto'] == 'Produto_A']
soma_Produto_A = total_venda_produto_A["Price"].sum()
display(total_venda_produto_A)
print(f"O valor Total de Vendas do Produto A foi: R$ {soma_Produto_A:.2f}")
```
