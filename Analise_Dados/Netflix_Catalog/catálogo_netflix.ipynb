```python
# Filtrar apenas os filmes ("Movie")
unique_movies_df = unique_movies_df[unique_movies_df['type'] == 'Movie']

# Categorizar os filmes com base na duração
unique_movies_df['duration_category'] = pd.cut(
    unique_movies_df['duration_min'],
    bins=[0, 60, 120, 180, float('inf')],
    labels=['Até 60 min', 'De 60 a 120 min', 'De 120 a 180 min', 'Acima de 180 min'],
    right=False
)

# Contar a quantidade de filmes em cada categoria
duration_counts = unique_movies_df['duration_category'].value_counts()

# Ordenar as categorias
duration_counts = duration_counts.reindex(['Até 60 min', 'De 60 a 120 min', 'De 120 a 180 min', 'Acima de 180 min'])

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
bars = duration_counts.plot(kind='bar', color='skyblue')
plt.title('Contagem de Filmes por Categoria de Duração')
plt.xlabel('Categoria de Duração')
plt.ylabel('Contagem de Filmes')
plt.xticks(rotation=0)

# Adicionar rótulos com a contagem acima de cada barra
for bar in bars.patches:
    plt.annotate(format(bar.get_height(), '.0f'),
                 (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                 ha='center', va='bottom')

plt.show()
```
